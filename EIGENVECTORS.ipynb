```python
import numpy as np
```

```python
B = np.array([[3,1],
             [0,2]])
print("Matrix B:")
print(B)
print()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(B)
```

```python
print("Eigenvalues:")
print(eigenvalues)
print()
```

```python
print("Eigenvectors:")
print(eigenvectors)
print()
```

```python
print("Verification of eigenvalue equation B*V = l*V:")
for i in range(len(eigenvalues)):
    Bv=np.dot(B, eigenvectors[:,1])
    lv=eigenvalues[1]*eigenvectors[:,1]
    print(f"\nverificationfor eigenvalue{eigenvalues[1]}:")
    print("B*V",Bv)
    print("B*V",lv)
```
